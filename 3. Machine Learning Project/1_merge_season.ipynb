```python
import pandas as pd
import glob
```

```python
df = pd.read_csv('data/done/apple.csv', encoding='cp949')
humidity = pd.read_csv('data/done/humidity.csv', encoding='cp949')
pressure = pd.read_csv('data/done/pressure.csv', encoding='cp949')
sun = pd.read_csv('data/done/sun.csv', encoding='cp949')
temp = pd.read_csv('data/done/temp.csv', encoding='cp949')
wind = pd.read_csv('data/done/wind_all.csv', encoding='cp949')
```

```python
season_humidity = humidity.copy()
season_humidity['계절'] = season_humidity['월'].replace([3,4,5,6,7,8,9], ['봄','봄','여름','여름','여름','가을','가을'])
season_humidity.drop(columns='월', inplace=True)

season_humidity = round(season_humidity.groupby(['년', '계절', '지역']).mean(), 1).reset_index()

spring_humidity = season_humidity[season_humidity['계절'] == '봄'].drop(columns='계절')
spring_humidity.columns = ['년', '지역', 'sp평균상대습도(%)','sp최소상대습도(%)']

summer_humidity = season_humidity[season_humidity['계절'] == '여름'].drop(columns='계절')
summer_humidity.columns = ['년', '지역', 'su평균상대습도(%)','su최소상대습도(%)']

fall_humidity = season_humidity[season_humidity['계절'] == '가을'].drop(columns='계절')
fall_humidity.columns = ['년', '지역', 'fa평균상대습도(%)','fa최소상대습도(%)']
```

```python
season_pressure = pressure.copy()
season_pressure['계절'] = season_pressure['월'].replace([3,4,5,6,7,8,9], ['봄','봄','여름','여름','여름','가을','가을'])
season_pressure.drop(columns='월', inplace=True)

season_pressure = round(season_pressure.groupby(['년', '계절', '지역']).mean(), 1).reset_index()

spring_pressure = season_pressure[season_pressure['계절'] == '봄'].drop(columns='계절')
spring_pressure.columns = ['년', '지역', 'sp평균현지기압(hPa)','sp평균해면기압(hPa)','sp최고해면기압(hPa)','sp최저해면기압(hPa)']

summer_pressure = season_pressure[season_pressure['계절'] == '여름'].drop(columns='계절')
summer_pressure.columns = ['년', '지역', 'su평균현지기압(hPa)','su평균해면기압(hPa)','su최고해면기압(hPa)','su최저해면기압(hPa)']

fall_pressure = season_pressure[season_pressure['계절'] == '가을'].drop(columns='계절')
fall_pressure.columns = ['년', '지역', 'fa평균현지기압(hPa)','fa평균해면기압(hPa)','fa최고해면기압(hPa)','fa최저해면기압(hPa)']
```

```python
season_sun = sun.copy()
season_sun['계절'] = season_sun['월'].replace([3,4,5,6,7,8,9], ['봄','봄','여름','여름','여름','가을','가을'])
season_sun.drop(columns='월', inplace=True)

season_sun = round(season_sun.groupby(['년', '계절', '지역']).mean(), 1).reset_index()

spring_sun = season_sun[season_sun['계절'] == '봄'].drop(columns='계절')
spring_sun.columns = ['년', '지역', 'sp합계일조시간(hr)','sp합계전천일사량(MJ/㎡)']

summer_sun = season_sun[season_sun['계절'] == '여름'].drop(columns='계절')
summer_sun.columns = ['년', '지역', 'su합계일조시간(hr)','su합계전천일사량(MJ/㎡)']

fall_sun = season_sun[season_sun['계절'] == '가을'].drop(columns='계절')
fall_sun.columns = ['년', '지역', 'fa합계일조시간(hr)','fa합계전천일사량(MJ/㎡)']
```

```python
season_temp = temp.copy()
season_temp['계절'] = season_temp['월'].replace([3,4,5,6,7,8,9], ['봄','봄','여름','여름','여름','가을','가을'])
season_temp.drop(columns='월', inplace=True)

season_temp = round(season_temp.groupby(['년', '계절', '지역']).mean(), 1).reset_index()

spring_temp = season_temp[season_temp['계절'] == '봄'].drop(columns='계절')
spring_temp.columns = ['년', '지역', 'sp평균기온(℃)', 'sp최고기온 평균(℃)', 'sp최저기온 평균(℃)', 'sp강수량(mm)']

summer_temp = season_temp[season_temp['계절'] == '여름'].drop(columns='계절')
summer_temp.columns = ['년', '지역', 'su평균기온(℃)', 'su최고기온 평균(℃)', 'su최저기온 평균(℃)', 'su강수량(mm)']

fall_temp = season_temp[season_temp['계절'] == '가을'].drop(columns='계절')
fall_temp.columns = ['년', '지역', 'fa평균기온(℃)', 'fa최고기온 평균(℃)', 'fa최저기온 평균(℃)', 'fa강수량(mm)']
```

```python
season_wind = wind.copy()
season_wind['계절'] = season_wind['월'].replace([3,4,5,6,7,8,9], ['봄','봄','여름','여름','여름','가을','가을'])
season_wind.drop(columns='월', inplace=True)

season_wind = round(season_wind.groupby(['년', '계절', '지역']).mean(), 1).reset_index()

spring_wind = season_wind[season_wind['계절'] == '봄'].drop(columns='계절')
spring_wind.columns = ['년', '지역', 'sp평균풍속(m/s)','sp최대풍속(m/s)']

summer_wind = season_wind[season_wind['계절'] == '여름'].drop(columns='계절')
summer_wind.columns = ['년', '지역', 'su평균풍속(m/s)','su최대풍속(m/s)']

fall_wind = season_wind[season_wind['계절'] == '가을'].drop(columns='계절')
fall_wind.columns = ['년', '지역', 'fa평균풍속(m/s)','fa최대풍속(m/s)']
```

```python
# weathers = []
seasons = [spring_humidity, summer_humidity, fall_humidity, spring_pressure, summer_pressure, fall_pressure, spring_sun, summer_sun, fall_sun, 
           spring_temp, summer_temp, fall_temp, spring_wind, summer_wind, fall_wind]
data = df.copy()
```
