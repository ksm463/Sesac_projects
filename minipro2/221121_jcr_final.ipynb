```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='United States')
```

```python
def geocoding(df_lat_loc):
    try:
        geo = geo_local.geocode(df_lat_loc)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
latitude = []
longitude = []

for i in df_lat_loc:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
# weapon_type 데이터를 가공할 함수 생성
def weapon_num(weapon_type) :
    num=0
    weapon_type_list = weapon_type.split(',')
    for weapon in weapon_type_list :
        if 'five' in weapon :
            num += 5
        elif 'three' in weapon :
            num += 3
        elif 'two' in weapon :
            num += 2
        else :
            num += 1
        
    return num
```

```python
# 적용, 임시로 생성한 컬럼 제거, 타입변환 
df3['weapon_type_test'] = df3['weapon_type'].str.lower().str.replace('handgun','HG').str.replace('and',',').str.replace(';',',').str.replace('2','two')
df3.loc[df3['numweapons'].isnull(),'numweapons'] = df3[df3['numweapons'].isnull()]['weapon_type_test'].apply(weapon_num)
df3 = df3.drop(columns='weapon_type_test')
df3 = df3.astype({'numweapons':'int'})
df3
```

```python
def find_group(value):
    groups = [10, 20, 30, 40, 50, 60, 70, 80]
    for i in range(len(groups)):
        if value+1 <= groups[i]:
            if i == 0:
                return 0
            
            return groups[i-1]
```
