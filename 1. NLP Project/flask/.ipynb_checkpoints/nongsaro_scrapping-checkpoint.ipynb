```python
import urllib.request
from bs4 import BeautifulSoup

apikey = '20230203TVSXGEZTOXMISORV8VWDA'  #'api키입력' #api 입력

#식물 이름과 컨텐츠 번호 추출하는 함수
def number_extraction(url):
    
    #식물 이름과 번호를 담을 공간
    plant_name  = []
    content_number = []
    
    #접속을 위한 url + key세팅 및 user-agent 설정
    header={'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Whale/3.19.166.16 Safari/537.36'}
    search_url= url+"?apiKey="+apikey+"&numOfRows=300"
    req=urllib.request.Request(url=search_url, headers=header)
    url_open=urllib.request.urlopen(req)
    beautifulsoup=BeautifulSoup(url_open,'html.parser')

    #식물 이름 검색(속성명 : cntntssj) -> 줄단위 텍스트 변환
    find_name = beautifulsoup.findAll('cntntssj')
    for i in find_name:
        plant_name.append(str(i.get_text()))
  
    #컨텐츠 번호 검색 (속성명 : cntntsno) -> 줄단위 텍스트 변환 -> 번호 이외에 문자 제거
    find_number = beautifulsoup.findAll('cntntsno')
    for i in find_number:
        content_number.append(str(i.get_text()).replace('<![CDATA[','').replace(']]>',''))
    
    # 변환 작업 후 Return
    return plant_name, content_number
       
#건조에 강한 식물
#dry_plant_name, dry_content_number = number_extraction('http://api.nongsaro.go.kr/service/dryGarden/dryGardenList')
# 잎관상 식물
leaves_plant_name , leaves_content_number = number_extraction('http://api.nongsaro.go.kr/service/garden/gardenList')


print(leaves_content_number)
```

```python
#컨텐츠 번호로 상세 검색하여 데이터 추출하는 함수
def detailed_search(url,plant_name,content_number,columns):
    
    #모든 데이터를 담을 공간
    all_data=[]
    
    #컨텐츠 번호로 조회하기 위한 반복문
    for i in range(len(content_number)):
        
        box=[]
        #식물명 삽입
        box.append(plant_name[i])
        
        #접속을 위한 url + key세팅 및 user-agent 설정
        header={'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Whale/3.19.166.16 Safari/537.36'}
        search_url=url+"?apiKey="+apikey+"&cntntsNo="+content_number[i]
        print(content_number[i],"=======================","i",i)
        req=urllib.request.Request(url=search_url, headers=header)
        url_open=urllib.request.urlopen(req)
        bsObject=BeautifulSoup(url_open,'html.parser')
        
        #속성을 지정한 컬럼으로 검색 -> 텍스트 처리 -> 필요없는 부분 제거
        for k in columns:
            box.append(bsObject.find(k).text.replace('<![CDATA','').replace(']>',''))
            
            #컬럼으로 검색된 데이터 같은 행에 삽입
        
        #이미지 가져오기
        box.append("https://www.nongsaro.go.kr/cms_contents/301/"+content_number[i]+"_MF_REPR_ATTACH_01.jpg")
        
        #box 데이터 줄 단위 삽입
        all_data.append(box)
        
    #모든 데이터 리턴
    return all_data

# dry_columns = ['cntntsno','stlesenm','tipinfo', 'managelevelnm', 'managedemandnm' , 'grwtsevenm', 'grwhtpinfo','hgtmmhmrinfo','frtlzrinfo','batchplaceinfo','dlthtsinfo','lighttinfo','watercycleinfo']
#leaves_columns = ['cntntsno','clcodenm','adviseinfo', 'managelevelcodenm', 'managedemanddocodenm' , 'grwtvecodenm', 'grwhtpcodenm','hdcodenm','frtlzrinfo','postngplacecodenm','dlthtscodenm','lighttdemanddocodenm','watercyclesprngcodenm']
# dry_plant = detailed_search('http://api.nongsaro.go.kr/service/dryGarden/dryGardenDtl',dry_plant_name, dry_content_number, dry_columns) 

leaves_columns = ['cntntsno',
    'managelevelcodenm',
    'managedemanddocodenm',
    'lighttdemanddocodenm',
    'postngplacecodenm',
    'grwtvecodenm',
    'grwhtpcodenm',
    'winterlwettpcodenm',
    'hdcodenm',
    'frtlzrinfo',
    'soilinfo',
    'watercyclesprngcodenm',
    'watercyclesummercodenm',
    'watercycleautumncodenm',
    'watercyclewintercodenm',
    'dlthtsmanageinfo']

leaves_plant = detailed_search('http://api.nongsaro.go.kr/service/garden/gardenDtl',leaves_plant_name, leaves_content_number, leaves_columns)

#combined_data = dry_plant+leaves_plant
```
